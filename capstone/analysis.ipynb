```python
import csv
import matplotlib.pyplot as plt

with open('portfolio.csv', 'r') as f:
    rows = csv.reader(f)
    headers = next(rows)
    my_dates = []
    prices = []
    for row in rows:
        my_dates.append(row[0])
        prices.append(float(row[3]))
        
plt.plot(my_dates, prices, color='green', linestyle='solid')

with open('portfolio2.csv', 'r') as f:
    rows = csv.reader(f)
    headers = next(rows)
    my_dates = []
    prices = []
    for row in rows:
        my_dates.append(row[0])
        prices.append(float(row[3]))

plt.plot(my_dates, prices, color='red', linestyle='solid')

with open('portfolio3.csv', 'r') as f:
    rows = csv.reader(f)
    headers = next(rows)
    my_dates = []
    prices = []
    for row in rows:
        my_dates.append(row[0])
        prices.append(float(row[3]))
        
plt.plot(my_dates, prices, color='blue', linestyle='solid')

with open('portfolio4.csv', 'r') as f:
    rows = csv.reader(f)
    headers = next(rows)
    my_dates = []
    prices = []
    for row in rows:
        my_dates.append(row[0])
        prices.append(float(row[3]))

plt.plot(my_dates, prices, color='orange', linestyle='solid')

plt.show()
```
